```python
def isHappy(number):
  history = set()
  while number != 1 and number not in history: #넘버가 1이 아니고 이전에 나온적이 없을 동안 반복
      history.add(number)                 #넘버를 문자열로써 추가한다
      digits = [int(d) for d in str(number)] #자릿수하나씩 뽑아서 정수로 바꾸고
      number = sum(d ** 2 for d in digits) # 각 자릿수 제곱한걸 다 더해서 다음숫자로
  return number == 1

print(isHappy(19))
```
